```python
import requests
from datetime import datetime

def get_apod():
    url = 'https://api.nasa.gov/planetary/apod?api_key=9AJFnzj6N5HxwD4iDReBXBsENwA7BLUHEfWYDidf'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        apod_url = data['url']
        explanation = data['explanation']
        print(f'\nСсылка на APOD: {apod_url}')
        print(f'Пояснение: {explanation}')

    else:
        print('Ошибка получения данных APOD')


def mars_rover_photos(date, camera):
    url = f'https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?earth_date={date}&camera={camera}&api_key=9AJFnzj6N5HxwD4iDReBXBsENwA7BLUHEfWYDidf'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        photos = data['photos']

        if photos:
            print('\nСписок фотографий:')
            for count, photo in enumerate(photos, 1):
                img_src = photo['img_src']
                print(f'{count}) {img_src}')
        else:
            print('\nФото не найдены')

    else:
        print('\nОшибка получения фотографий с марсохода')


def get_neo_feed(start_date, end_date):
    url = f'https://api.nasa.gov/neo/rest/v1/feed?start_date={start_date}&end_date={end_date}&api_key=9AJFnzj6N5HxwD4iDReBXBsENwA7BLUHEfWYDidf'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        neo_data = data['near_earth_objects']

        for date, objects in neo_data.items():
            print(f'\nДата: {date}')
            for obj in objects:
                name = obj['name']
                diameter = obj['estimated_diameter']['meters']['estimated_diameter_max']
                hazardous = obj['is_potentially_hazardous_asteroid']
                print(f'Название: {name}, Диаметр (м): {diameter}, Потенциально опасный: {hazardous}')
    else:
        print('Ошибка получения данных об объектах, сближающихся с Землей')


def get_donki(startDate, endDate):
    #Геомагнитная буря
    url_gst = f'https://api.nasa.gov/DONKI/GST?startDate={startDate}&endDate={endDate}&api_key=9AJFnzj6N5HxwD4iDReBXBsENwA7BLUHEfWYDidf'
    #Солнечная вспышка
    url_flr = f'https://api.nasa.gov/DONKI/FLR?startDate={startDate}&endDate={endDate}&api_key=9AJFnzj6N5HxwD4iDReBXBsENwA7BLUHEfWYDidf'

    response_gst = requests.get(url_gst)
    response_flr = requests.get(url_flr)

    if response_gst.status_code == 200 and response_flr.status_code == 200:
        data_gst = response_gst.json()
        data_flr = response_flr.json()

        print('Данные о геомагнитной буре:')
        for event in data_gst:
            print(f"Начало: {event['startTime']}")

        print('\nДанные о солнечной вспышке:')
        for event in data_flr:
             print(f'Дата: {event["beginTime"]}, Уровень: {event["classType"]}')

    else:
        print('Ошибка получения данных о космической погоде')


def main():
    while True:
        functions = ['Получение астрономической картинки дня APOD',
                    'Получение фотографий с марсохода по дате и типу камеры',
                    'Поиск и отображение информации об объекте, сближающихся с Землей',
                    'Получение данных о космической погоде', 'Выход из программы']

        print('\nФункции программы: ')
        for count, function in enumerate(functions, 1):
            print(f'{count}) {function}')

        command = int(input('\nВведите номер функции, которую хотите изучить: '))

        if command == 1:
            get_apod()

        elif command == 2:
            date = input('\nВведите дату (YYYY-MM-DD): ')
            try:
                datetime.strptime(date, '%Y-%m-%d')
            except ValueError:
                print('Неверный формат даты')
                continue

            type_cam = ['FHAZ', 'RHAZ' , 'MAST', 'CHEMCAM', 'MAHLI', 'MARDI',
                        'NAVCAM', 'PANCAM', 'MINITES']

            print('\nТипы камер: ')
            for count, type in enumerate(type_cam, 1):
                print(f'{count}) {type}')

            camera = input('\nВведите тип камеры: ')

            print(mars_rover_photos(date, camera))

        elif command == 3:
            start_date = input('\nВведите дату начала поиска (YYYY-MM-DD): ')
            try:
                datetime.strptime(start_date, '%Y-%m-%d')
            except ValueError:
                print('Неверный формат даты начала')
                continue

            end_date = input('\nВведите дату конца поиска (YYYY-MM-DD): ')
            try:
                datetime.strptime(end_date, '%Y-%m-%d')
            except ValueError:
                print('Неверный формат даты конца')
                continue

            print(get_neo_feed(start_date, end_date))

        elif command == 4:
            startDate = input('\nВведите дату начала поиска (YYYY-MM-DD): ')
            try:
                datetime.strptime(startDate, '%Y-%m-%d')
            except ValueError:
                print('Неверный формат даты начала')
                continue

            endDate = input('\nВведите дату конца поиска (YYYY-MM-DD): ')
            try:
                datetime.strptime(endDate, '%Y-%m-%d')
            except ValueError:
                print('Неверный формат даты конца')

            get_donki(startDate, endDate)

        elif command == 5:
            break

        else:
           print('Неверный номер функции')


main()



```

Цель этой задачи - создать скрипт на Python, который взаимодействует с API Чикагского института искусств (https://api.artic.edu/docs/) для извлечения и отображения произведений искусства. Скрипт должен позволять пользователям просматривать работы по страницам, фильтровать их по имени художника и просматривать подробную информацию о выбранных произведениях искусства. Ниже приведены требования и функциональные возможности, которые необходимо реализовать:

Требования:
Извлекать произведения искусства:

- Создайте функцию, которая извлекает список произведений искусства из API Чикагского института искусств.
Функция должна принимать параметр page для разбивки на страницы и возвращать список произведений искусства вместе с информацией о разбивке на страницы.
Фильтровать произведения искусства:

- Реализуйте функцию, которая фильтрует список произведений искусства на основе имени указанного художника. Функция должна возвращать список работ, которые соответствуют имени художника (без учета регистра).
Отображать подробную информацию об оформлении:

- Напишите функцию, которая отображает названия работ для пользователя и позволяет ему выбрать одну из них, введя соответствующий номер.
После выбора функция должна отображать подробную информацию о выбранном произведении, включая название, исполнителя, дату и носитель.
Разбивка на страницы и взаимодействие с пользователем:

- Создайте основную функцию, которая управляет выборкой произведений и взаимодействием с пользователем.

Разрешите пользователям перемещаться по страницам с произведениями искусства, выполнять фильтрацию по исполнителю или выходить из программы.

Если страниц с произведениями искусства несколько, укажите варианты перехода к следующей странице, предыдущей странице, фильтрации по исполнителю или выхода из программы.

```python
import requests

# URL API
API_URL = "https://api.artic.edu/api/v1/artworks"


def get_artworks(page=1):
    """
    Извлекает список произведений искусства из API.
    Возвращает список произведений и информацию о разбивке на страницы.
    """
    params = {
        "page": page,
        "limit": 10,  # кол-во произведений на странице
        "fields": "id,title,artist_title,date_display,medium_display"  # запрашиваемые поля
    }
    response = requests.get(API_URL, params=params)
    data = response.json()

    artworks = data["data"]
    pagination = data["pagination"]

    return artworks, pagination


def filter_artworks(all_artworks, artist_name):
    """
    Фильтрует список произведений по имени художника.
    Возвращает отфильтрованный список произведений.
    """
    filtered_artworks = []

    for artwork in all_artworks:
        artist_title = artwork.get("artist_title")
        if artist_title is not None and artist_name.lower() == artist_title.lower():
            filtered_artworks.append(artwork)

    return filtered_artworks


def display_artwork_details(artwork):
    """
    Отображает подробную информацию о произведении искусства.
    """
    print(f"\nНазвание: {artwork['title']}")
    print(f"Художник: {artwork['artist_title']}")
    print(f"Дата: {artwork['date_display']}")
    print(f"Носитель: {artwork['medium_display']}\n")


def main():
    current_page = 1
    artworks, pagination = get_artworks(current_page)
    original_artworks = artworks.copy()  # Копия исходного списка
    current_page_artworks = artworks.copy()  # Копия произведений текущей страницы

    while True:
        print(f"\nСтраница {current_page} из {pagination['total_pages']}")
        for i, artwork in enumerate(artworks, start=1):
            print(f"{i}. {artwork['title']} ({artwork['artist_title']})")

        choice = input("\nВведите 'c' для след. стр., 'p' для пред. стр., 'f' для фильтр. по художнику, 'r' для возвр. к исход. списку, 'd' для подроб. инф., или 'q' для выхода: ").lower()

        if choice == 'c':
            if current_page < pagination['total_pages']:
                current_page += 1
                artworks, pagination = get_artworks(current_page)
                current_page_artworks = artworks.copy()  # Обновление копии текущей страницы
            else:
                print("\nВы на последней странице.")

        elif choice == 'p':
            if current_page > 1:
                current_page -= 1
                artworks, pagination = get_artworks(current_page)
                current_page_artworks = artworks.copy()  # Обновление копии текущей страницы
            else:
                print("\nВы на первой странице.")

        elif choice == 'f':
            artist_name = input("\nВведите имя художника: ")
            filtered_artworks = filter_artworks(current_page_artworks, artist_name)

            if filtered_artworks:
                artworks = filtered_artworks
                current_page = 1  # Сбрасываем текущую страницу на 1
                pagination['total_pages'] = 1  # Только одна страница для отфильтрованных
            else:
                print("\nПроизведения не найдены.")

        elif choice == 'r':
            artworks = original_artworks  # Возврат к исходному списку
            pagination = get_artworks(current_page)[1]  # Обновляем информацию о разбивке на страницы

        elif choice == 'd':
            artwork_index = int(input("\nВведите номер произведения: "))

            if 1 <= artwork_index <= len(artworks):
                display_artwork_details(artworks[artwork_index - 1])
            else:
                print("\nНекорректный номер произведения.")

        elif choice == 'q':
            print("\nВыход из программы.")
            break

        else:
            print("\nНекорректный ввод. Пожалуйста, попробуйте снова.")


main()
```

# Задание 3

Задача: Создать программу по управлению портфелем криптовалют

Цель: Создать скрипт на Python, который извлекает цены на криптовалюты в режиме реального времени, позволяет пользователям управлять портфелем криптовалют, вычисляет общую стоимость портфеля, отслеживает изменения цен и предоставляет исторические данные о ценах для анализа.

Требования:
Получение текущих цен на криптовалюты:

Используйте https://docs.coingecko.com/ для получения актуальных цен на список криптовалют.

Управление портфелем:

- Позволяет пользователю создавать портфель криптовалют и управлять им, указывая количество каждой криптовалюты, которой он владеет.
- Расчитывает общую стоимость портфеля в указанной фиатной валюте (например, долларах США).

Отслеживание изменения цен:

- Отображение процентного изменения цены для каждой криптовалюты в портфеле за последние 24 часа.
- Выделите все криптовалюты, стоимость которых значительно увеличилась или снизилась.

Поиск исторических данных о ценах:

- Получение исторических данных о ценах на указанную криптовалюту за последнюю неделю.
- Предоставьте пользователю возможность визуализировать эти данные в простом текстовом формате (например, цены за день).

Взаимодействие с пользователем:

- Реализуйте интерфейс командной строки для ввода данных пользователем.
- Предоставьте опции для получения текущих цен, управления портфелем, просмотра изменений цен или анализа исторических данных.


```python
import requests
from datetime import datetime


api_base_url = "https://api.coingecko.com/api/v3"
portfolio = {}


def get_crypto_price(crypto_id, currency='usd'):
    """Получение текущей цены криптовалюты"""
    url = f"{api_base_url}/simple/price"
    params = {
        'ids': crypto_id,
        'vs_currencies': currency,
        'include_24hr_change': 'true'
    }

    response = requests.get(url, params=params)
    return response.json().get(crypto_id)


def add_to_portfolio(crypto_id, amount):
    """Добавление криптовалюты в портфель"""
    portfolio[crypto_id] = amount
    print(f"Добавлено {amount} {crypto_id} в портфель")


def remove_from_portfolio(crypto_id):
    """Удаление криптовалюты из портфеля"""
    if crypto_id in portfolio:
        del portfolio[crypto_id]
        print(f"Удалено {crypto_id} из портфеля")
    else:
        print(f"{crypto_id} не найден в портфеле")


def get_portfolio_value(currency='usd'):
    """Расчет общей стоимости портфеля"""
    total_value = 0
    print("\nСводка портфеля:")
    print("-----------------")

    for crypto_id, amount in portfolio.items():
        price_data = get_crypto_price(crypto_id, currency)

        if price_data:
            current_price = price_data[currency]
            value = amount * current_price
            change_24h = price_data.get(f'{currency}_24h_change', 0)

            total_value += value
            print(f"{crypto_id.upper()}:")
            print(f"Количество: {amount}")
            print(f"Текущая цена: ${current_price:.2f}")
            print(f"Общая стоимость: ${value:.2f}")
            print(f"Изменение за 24 часа: {change_24h:.2f}%")
            print("-----------------")

    print(f"\nОбщая стоимость портфеля: ${total_value:.2f}")
    return total_value


def get_historical_data(crypto_id, days=7):
    """Получение исторических данных"""
    url = f"{api_base_url}/coins/{crypto_id}/market_chart"
    params = {'vs_currency': 'usd', 'days': days}
    response = requests.get(url, params=params)
    prices = response.json().get('prices', [])

    print(f"\nИсторические цены для {crypto_id.upper()} (за последние {days} дней):")
    for timestamp, price in prices:
        date = datetime.fromtimestamp(timestamp / 1000).strftime('%Y-%m-%d')
        print(f"{date}: ${price:.2f}")


def main():
    while True:
        print("\nМенеджер криптовалютного портфеля")
        print("1. Добавить криптовалюту в портфель")
        print("2. Удалить криптовалюту из портфеля")
        print("3. Показать стоимость портфеля")
        print("4. Показать исторические данные")
        print("5. Выйти")

        choice = input("\nВведите ваш выбор (1-5): ")

        if choice == '1':
            crypto_id = input("\nВведите идентификатор криптовалюты (например, bitcoin): ").lower()
            amount = float(input("Введите количество: "))
            add_to_portfolio(crypto_id, amount)

        elif choice == '2':
            crypto_id = input("\nВведите идентификатор криптовалюты для удаления: ").lower()
            remove_from_portfolio(crypto_id)

        elif choice == '3':
            get_portfolio_value()

        elif choice == '4':
            crypto_id = input("\nВведите идентификатор криптовалюты: ").lower()
            days = int(input("Введите количество дней: "))
            get_historical_data(crypto_id, days)

        elif choice == '5':
            print("\nДо свидания!")
            break

        else:
            print("\nНекорректный выбор. Пожалуйста, попробуйте снова.")


main()
```
