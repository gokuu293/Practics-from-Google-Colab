# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary


class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_salary(self):
        return self.base_salary + self.bonus


class Developer(Employee):
    def __init__(self, name, base_salary, hours_worked):
        super().__init__(name, base_salary)
        self.hours_worked = hours_worked

    def calculate_salary(self):
        overtime_pay = 0
        if self.hours_worked > 160:
            overtime_hours = self.hours_worked - 160
            overtime_pay = overtime_hours * (self.base_salary / 160)
        return self.base_salary + overtime_pay


employee = Employee('Александр', 5000)
manager = Manager('Борис', 5000, 2000)
developer = Developer('Вячеслав', 5000, 240)


print(f'Зарплата сотрудника по имени {employee.name}: {employee.calculate_salary()}')
print(f'Зарплата менеджера по имени {manager.name}: {manager.calculate_salary()}')
print(f'Зарплата разработчика имени {developer.name}: {developer.calculate_salary()}')
```

# Задание 3

Онлайн-магазин:
- Создайте модель для онлайн-магазина с классами Product, Order, Customer, и ShoppingCart.
- Product включает информацию о цене, наличии на складе и категории товара.
Order обрабатывает процесс покупки, включая расчет цены с учетом скидок и налогов.
- Customer управляет информацией о пользователе и его истории заказов.
- ShoppingCart позволяет добавлять, удалять и обновлять количество товаров перед оформлением заказа.


```python
class Product:
    def __init__(self, name, price, stock, category):
        # name - название товара
        # price - цена за единицу
        # stock - количество товара на складе
        # category - категория товара (например, "Электроника", "Одежда")
        self.name = name
        self.price = price
        self.stock = stock
        self.category = category

class Customer:
    def __init__(self, name):
        # name - имя покупателя
        # order_history - список предыдущих заказов данного покупателя
        self.name = name
        self.order_history = []

class ShoppingCart:
    def __init__(self):
        # items - словарь, где ключ - объект Product, а значение - количество
        self.items = {}

    def add_product(self, product, quantity=1):
        # Добавляем товар в корзину в указанном количестве
        # Проверяем, достаточно ли товара на складе
        if product.stock < quantity:
            print(f"Недостаточно товара {product.name} на складе.")
            return
        # Если товар уже есть в корзине, увеличиваем количество
        self.items[product] = self.items.get(product, 0) + quantity

    def remove_product(self, product, quantity=1):
        # Удаляем товар из корзины или уменьшаем его количество
        if product not in self.items:
            print(f"Товар {product.name} отсутствует в корзине.")
            return
        current_qty = self.items[product]
        if quantity >= current_qty:
            # Если удаляемое количество больше или равно имеющемуся, удаляем товар полностью
            del self.items[product]
        else:
            # Иначе просто уменьшаем количество
            self.items[product] = current_qty - quantity

    def update_product_quantity(self, product, quantity):
        # Обновляем количество товара в корзине до указанного значения
        if product not in self.items:
            print(f"Товар {product.name} отсутствует в корзине.")
            return
        if product.stock < quantity:
            print(f"Недостаточно товара {product.name} на складе.")
            return
        if quantity <= 0:
            # Если количество меньше или равно нулю, товар убирается из корзины
            del self.items[product]
        else:
            self.items[product] = quantity

    def calculate_total(self):
        # Подсчет итоговой суммы всех товаров в корзине без учета скидок и налогов
        total = 0
        for product, qty in self.items.items():
            total += product.price * qty
        return total

class Order:
    def __init__(self, customer, cart, discount=0, tax_rate=0.2):
        # customer - объект покупателя
        # cart - корзина, из которой будет оформлен заказ
        # discount - скидка в виде десятичной дроби (например, 0.1 = 10%)
        # tax_rate - налоговая ставка в виде десятичной дроби (например, 0.2 = 20%)
        self.customer = customer
        self.cart = cart
        self.discount = discount
        self.tax_rate = tax_rate

        # Рассчитываем итоговую цену заказа с учетом скидки и налогов
        self.total = self.calculate_total_price()

        # Завершаем оформление заказа: обновляем остатки и историю заказов
        self.finalize_order()

    def calculate_total_price(self):
        # Сначала получаем сумму без скидок и налогов
        subtotal = self.cart.calculate_total()
        # Применяем скидку
        subtotal_after_discount = subtotal * (1 - self.discount)
        # Применяем налог
        total_with_tax = subtotal_after_discount * (1 + self.tax_rate)
        return total_with_tax

    def finalize_order(self):
        # Уменьшаем количество товаров на складе согласно заказу
        for product, qty in self.cart.items.items():
            product.stock -= qty
        # Добавляем заказ в историю покупателя
        self.customer.order_history.append(self)

def print_products(products):
    # Выводим список товаров с ценой, остатком и категорией
    print("\nДоступные товары:")
    for i, product in enumerate(products, start=1):
        print(f"{i}. {product.name} - Цена: {product.price} руб, Остаток: {product.stock}, Категория: {product.category}")

def print_cart(cart):
    # Выводим содержимое корзины
    if not cart.items:
        print("\nВаша корзина пуста.")
        return
    print("\nВаша корзина:")
    for product, qty in cart.items.items():
        print(f"{product.name} (x{qty}) - Сумма: {product.price * qty} руб")
    print(f"Итого: {cart.calculate_total()} руб")


def main():
    # Создаем примерные товары
    products = [
        Product("Ноутбук", 50000, 10, "Электроника"),
        Product("Телефон", 30000, 5, "Электроника"),
        Product("Футболка", 1000, 20, "Одежда")
    ]

    # Создаем покупателя
    customer_name = input("Введите ваше имя: ")
    customer = Customer(customer_name)

    # Создаем корзину
    cart = ShoppingCart()

    while True:
        print("\nГлавное меню:")
        print("1. Просмотреть товары")
        print("2. Добавить товар в корзину")
        print("3. Удалить товар из корзины")
        print("4. Посмотреть корзину")
        print("5. Оформить заказ")
        print("6. Посмотреть историю заказов")
        print("7. Выйти")

        choice = input("Выберите пункт меню: ")

        if choice == "1":
            print_products(products)
        elif choice == "2":
            print_products(products)
            try:
                product_choice = int(input("Введите номер товара: ")) - 1
                if 0 <= product_choice < len(products):
                    product = products[product_choice]
                    quantity = int(input(f"Сколько '{product.name}' хотите добавить? "))
                    cart.add_product(product, quantity)
                    print(f"Товар '{product.name}' добавлен в корзину в количестве {quantity} шт.")
                else:
                    print("Неверный номер товара.")
            except ValueError:
                print("Пожалуйста, введите корректное число.")
        elif choice == "3":
            if not cart.items:
                print("Корзина пуста.")
            else:
                print_cart(cart)
                product_names = [p.name for p in cart.items.keys()]
                product_name_to_remove = input("Введите название товара, который хотите удалить: ")
                # Находим товар по имени
                product_to_remove = None
                for p in cart.items.keys():
                    if p.name == product_name_to_remove:
                        product_to_remove = p
                        break
                if product_to_remove is None:
                    print("Такого товара нет в корзине.")
                else:
                    qty_to_remove = int(input("Сколько штук удалить (или введите большое число для полного удаления): "))
                    cart.remove_product(product_to_remove, qty_to_remove)
                    print(f"Товар '{product_to_remove.name}' удален/уменьшен в корзине.")
        elif choice == "4":
            print_cart(cart)
        elif choice == "5":
            if not cart.items:
                print("Нельзя оформить заказ. Корзина пуста.")
            else:
                # Простая логика оформления заказа,
                # можно ввести скидку и налог вручную или использовать значения по умолчанию
                discount_str = input("Введите скидку в %, например '10' или оставьте пустым: ")
                tax_str = input("Введите налог в %, например '20' или оставьте пустым: ")

                try:
                    discount = float(discount_str)/100 if discount_str else 0
                except:
                    discount = 0

                try:
                    tax_rate = float(tax_str)/100 if tax_str else 0.2
                except:
                    tax_rate = 0.2

                order = Order(customer, cart, discount=discount, tax_rate=tax_rate)
                print(f"Заказ оформлен. Итоговая сумма: {order.total:.2f} руб.")
                # Очищаем корзину
                cart = ShoppingCart()
        elif choice == "6":
            if not customer.order_history:
                print("У вас еще нет истории заказов.")
            else:
                print("\nИстория заказов:")
                for i, order in enumerate(customer.order_history, start=1):
                    print(f"Заказ №{i}: Сумма {order.total:.2f} руб")
        elif choice == "7":
            print("Спасибо за использование онлайн-магазина!")
            break
        else:
            print("Неверный выбор. Попробуйте еще раз.")

if __name__ == "__main__":
    main()
```

# Задание 4

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
import random

class SpaceShip:
    def __init__(self, name, fuel=100, hull=100, speed=0):
        # Инициализирует космический корабль с начальными параметрами:
        # name: название корабля (строка)
        # fuel: количество топлива (число)
        # hull: состояние корпуса (число)
        # speed: текущая скорость (число)
        self.name = name
        self.fuel = fuel
        self.hull = hull
        self.speed = speed

    def accelerate(self, amount):
        # Увеличивает скорость корабля на заданное значение amount.
        # Расход топлива равен величине ускорения.
        # Если топлива недостаточно, ускорение не происходит.
        if self.fuel > amount:
            self.speed += amount
            self.fuel -= amount
            print(f"{self.name}: Ускорение на {amount}, скорость теперь {self.speed}, топливо {self.fuel}.")
        else:
            print(f"{self.name}: Недостаточно топлива для ускорения.")

    def decelerate(self, amount):
        # Уменьшает скорость корабля на заданное значение amount.
        # Скорость не может стать меньше 0.
        self.speed = max(0, self.speed - amount)
        print(f"{self.name}: Торможение на {amount}, скорость теперь {self.speed}.")

    def take_damage(self, amount):
        # Наносит кораблю урон по корпусу на указанную величину amount.
        # Если прочность корпуса падает до 0 или ниже, корабль считается уничтоженным.
        self.hull -= amount
        if self.hull <= 0:
            print(f"{self.name}: Корпус разрушен! Корабль уничтожен.")
        else:
            print(f"{self.name}: Корпус поврежден на {amount}, прочность корпуса {self.hull}.")

    def refuel(self, amount):
        # Изменяет количество топлива на заданную величину (может быть отрицательной, если расход).
        self.fuel += amount
        print(f"{self.name}: Заправка на {amount}, топливо теперь {self.fuel}.")


class CrewMember:
    def __init__(self, name, role, health=100, skill=50):
        # Инициализирует члена экипажа.
        # name: имя члена экипажа
        # role: роль в экипаже (например, "пилот", "инженер")
        # health: здоровье члена экипажа
        # skill: навыки, число определяющее мастерство/эффективность
        self.name = name
        self.role = role
        self.health = health
        self.skill = skill

    def work(self, spaceship):
        # Действия члена экипажа.
        # Если инженер, пытается починить корпус, если он не полон.
        # Если пилот, может снижать расход топлива (упрощено до вывода сообщения).
        # Если роль не предусмотрена, выводится сообщение о бездействии.
        if self.role == "инженер" and spaceship.hull < 100:
            # Чиним корпус на 10 единиц или до полного восстановления (макс 100)
            repair_amount = min(10, 100 - spaceship.hull)
            spaceship.hull += repair_amount
            print(f"{self.name} (инженер) починил корпус на {repair_amount}, прочность корпуса: {spaceship.hull}")
        elif self.role == "пилот" and spaceship.fuel > 0:
            # Пилот улучшает маневрирование (здесь просто сообщение)
            print(f"{self.name} (пилот) улучшил маневрирование.")
        else:
            print(f"{self.name} ({self.role}) ничего полезного не сделал(а).")

    def take_damage(self, amount):
        # Член экипажа получает урон по здоровью.
        # Если здоровье падает до 0 или ниже, член экипажа погибает.
        self.health -= amount
        if self.health <= 0:
            print(f"{self.name}: погиб.")
        else:
            print(f"{self.name}: получил урон {amount}, здоровье: {self.health}.")


class Mission:
    def __init__(self, spaceship, crew_members, objectives):
        # Инициализирует миссию.
        # spaceship: объект класса SpaceShip
        # crew_members: список объектов класса CrewMember
        # objectives: список целей миссии (строки)
        self.spaceship = spaceship
        self.crew_members = crew_members
        self.objectives = objectives
        # Возможные события, которые могут произойти во время миссии
        self.events = ["авария", "метеорит", "затишье", "авария двигателя"]

    def run(self):
        # Запускает миссию.
        # Для каждой цели случайно выбирается событие.
        # В зависимости от события, корабль и/или экипаж получают урон или теряют ресурсы.
        # По окончании проверки всех целей, если корабль не уничтожен, миссия считается выполненной.
        print("Миссия началась!")
        for obj in self.objectives:
            print(f"Цель: {obj}")

            # Выбираем случайное событие
            event = random.choice(self.events)

            if event == "авария":
                print("Случилось аварийное событие!")
                damage = random.randint(5, 20)
                self.spaceship.take_damage(damage)
                # Один из членов экипажа получает урон
                injured_member = random.choice(self.crew_members)
                injured_member.take_damage(damage // 2)

            elif event == "метеорит":
                print("Столкновение с метеоритом!")
                self.spaceship.take_damage(random.randint(10, 30))

            elif event == "авария двигателя":
                print("Сбой двигателя, потребуется дополнительное топливо!")
                self.spaceship.refuel(-random.randint(10, 30))

            else:
                print("Полет проходит спокойно.")

            # После события члены экипажа пытаются улучшить ситуацию
            for member in self.crew_members:
                member.work(self.spaceship)

            # Проверяем состояние корабля после всех действий
            if self.spaceship.hull <= 0:
                print("Миссия провалена!")
                return

        # Если все цели достигнуты и корабль не уничтожен
        print("Все цели достигнуты! Миссия завершена.")


# Пример использования кода
if __name__ == "__main__":
    ship = SpaceShip("Звездный скиталец")
    crew = [
        CrewMember("Андрей", "пилот"),
        CrewMember("Елена", "инженер"),
    ]
    mission = Mission(ship, crew, ["Добраться до орбиты Марса", "Разведать астероидное поле"])
    mission.run()
```

```python
class Board:
    def __init__(self):
        # Изначально поле — это список из 9 пробелов (3x3)
        self.cells = [" " for _ in range(9)]

    def display(self):
        # Отображаем игровое поле в формате:
        #  1 | 2 | 3
        # ---+---+---
        #  4 | 5 | 6
        # ---+---+---
        #  7 | 8 | 9
        print(f" {self.cells[0]} | {self.cells[1]} | {self.cells[2]} ")
        print("---+---+---")
        print(f" {self.cells[3]} | {self.cells[4]} | {self.cells[5]} ")
        print("---+---+---")
        print(f" {self.cells[6]} | {self.cells[7]} | {self.cells[8]} ")

    def make_move(self, position, symbol):
        # position от 1 до 9, но в списке индексация с 0
        if self.is_cell_free(position):
            self.cells[position - 1] = symbol
            return True
        return False

    def is_cell_free(self, position):
        return self.cells[position - 1] == " "

    def check_winner(self, symbol):
        # Проверка всех комбинаций для победы
        wins = [
            [0, 1, 2],  # первая строка
            [3, 4, 5],  # вторая строка
            [6, 7, 8],  # третья строка
            [0, 3, 6],  # первый столбец
            [1, 4, 7],  # второй столбец
            [2, 5, 8],  # третий столбец
            [0, 4, 8],  # диагональ слева направо
            [2, 4, 6]   # диагональ справа налево
        ]

        for combo in wins:
            if self.cells[combo[0]] == symbol and self.cells[combo[1]] == symbol and self.cells[combo[2]] == symbol:
                return True
        return False

    def is_full(self):
        # Если нет свободных ячеек, то поле заполнено.
        return all(cell != " " for cell in self.cells)


class Player:
    def __init__(self, name, symbol):
        self.name = name
        self.symbol = symbol

    def make_move(self, board):
        # Запрашиваем ввод хода у игрока
        while True:
            try:
                move = int(input(f"{self.name} ({self.symbol}), введите номер клетки (1-9): "))
                if move < 1 or move > 9:
                    print("Введите число от 1 до 9!")
                    continue
                if board.make_move(move, self.symbol):
                    break
                else:
                    print("Клетка занята! Попробуйте снова.")
            except ValueError:
                print("Некорректный ввод! Введите число от 1 до 9.")


class Game:
    def __init__(self, player1, player2):
        self.board = Board()
        self.current_player = player1
        self.other_player = player2

    def switch_player(self):
        self.current_player, self.other_player = self.other_player, self.current_player

    def start(self):
        # Основной игровой цикл
        while True:
            self.board.display()
            self.current_player.make_move(self.board)

            # Проверяем победителя
            if self.board.check_winner(self.current_player.symbol):
                self.board.display()
                print(f"Поздравляем, {self.current_player.name}! Вы победили.")
                break

            # Проверяем на ничью
            if self.board.is_full():
                self.board.display()
                print("Ничья! Поле заполнено.")
                break

            # Переход хода
            self.switch_player()


# Инициализация игры
if __name__ == "__main__":
    p1 = Player("Игрок 1", "X")
    p2 = Player("Игрок 2", "O")

    game = Game(p1, p2)
    game.start()
```
