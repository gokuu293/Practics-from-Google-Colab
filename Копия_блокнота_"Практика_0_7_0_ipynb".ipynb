## Задание 1. HTTP-запросы, ответы и погода

Описание:

Напишите HTTP-запрос для получения информации о погоде в введенном городе из API.

Можно использовать API: https://open-meteo.com/. Используйте метод GET.


Ввод
```
56.50, 60.35
```

Вывод
```
Сегодня (1.11) погода 20 ◦С, нет осадков, туман
```





````python
import requests


count_temp = sum_temp = 0
count_cloud = sum_cloud = 0
check_fog = 0

data = input('').split(', ')

latitude = data[0]

longitude = data[1]

url = f'https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&hourly=temperature_2m,precipitation_probability,cloud_cover_low&daily=precipitation_probability_max&forecast_days=1'

responce = requests.get(url)
json = responce.json()
#Дата
date = json['daily']['time'][0]
#Вероятность выпадения осадков в %
precipitation = json['daily']['precipitation_probability_max'][0]
#Почасовая температура
temperatures = json['hourly']['temperature_2m']
#Низкая облачность и туман на высоте до 3 км в %
cloud_cover_low = json['hourly']['cloud_cover_low']
#Средняя температура за день
for temperature in temperatures:
  count_temp += 1
  sum_temp += temperature

avg_temp = int(sum_temp // count_temp)
#Низкая облачность и туман на высоте до 3 км в % за день
for cloud in cloud_cover_low:
  count_cloud += 1
  sum_cloud += cloud

avg_cloud = int(sum_cloud//count_cloud)

if avg_cloud > 80:
  check_fog = 'Туман'
else:
  check_fog = 'Тумана нет'

month = date[5:7]
day = date[8:10]
date = f'({day}.{month})'

print(f'Сегодня {date} погода {avg_temp} ◦С, вероятность осадков {precipitation} %, {check_fog} ')
````

````python
import requests


# 1) Получить список первых 20 покемонов
url = "https://pokeapi.co/api/v2/pokemon?limit=20"
response = requests.get(url)
data = response.json()

# 2) Извлечь и вывести имена покемонов
pokemon_names = [pokemon["name"] for pokemon in data["results"]]
print("Список первых 20 покемонов:")
print("\n".join(pokemon_names))

# 3) Ввести название покемона
pokemon_name = input("Имя покемона: ")

# 4) Получить информацию о выбранном покемоне
pokemon_url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
response = requests.get(pokemon_url)
pokemon_data = response.json()

# 5) Вывести информацию о выбранном покемоне
print("\nИнформация о покемоне:")
print(f"Имя: {pokemon_data['name']}")
print(f"Тип: {', '.join(type['type']['name'] for type in pokemon_data['types'])}")
print(f"Вес: {pokemon_data['weight']} кг")
print(f"Рост: {pokemon_data['height']} м")
print("Способности:", ", ".join(ability['ability']['name'] for ability in pokemon_data['abilities']))
````

## Задание 3. HTTP-запросы, ответы и посты

**Описание:**

Создайте программу, которая будет взаимодействовать с JSONPlaceholder API, реализуя следующие функции:

1. Реализуйте функцию, которая выполняет GET-запрос к https://jsonplaceholder.typicode.com/posts и возвращает список постов в формате JSON

2. Реализуйте функцию, котороая получает вводимое ID поста, выполняет GET-запрос по ID и возвращает данные поста в формате JSON

3. Реализуйте функцию, которая выполняет обработку JSON из пункта 2 и выводит всю важную информацию в консоль

````python
import requests


# 1) Функция для получения списка всех постов
def get_all_posts():
    response = requests.get("https://jsonplaceholder.typicode.com/posts")
    posts = response.json()
    return posts

# 2) Функция для получения поста по ID
def get_post_by_id(post_id):
    url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
    response = requests.get(url)
    post = response.json()
    return post

# 3) Функция для вывода информации о посте
def print_post_info(post):
    print(f"ID: {post['id']}")
    print(f"Заголовок: {post['title']}")
    print(f"Текст: {post['body']}")
    print(f"Идентификатор пользователя: {post['userId']}")

# Главная функция
def main():
    # Получить все посты
    all_posts = get_all_posts()
    print(f"Всего постов: {len(all_posts)}")

    # Получить ID поста от пользователя
    post_id = int(input("Введите ID поста: "))

    # Получить пост по ID
    post = get_post_by_id(post_id)

    # Вывести информацию о посте
    print("\nИнформация о посте:")
    print_post_info(post)

main()
````

## Задание 4. HTTP-запросы, ответы и работа с постами

**Описание**

Создайте программу, которая будет взаимодействовать с JSONPlaceholder API (из предыдущего задания), реализуя новые функции:

1. Реализуйте функцию, которая принимает заголовок, содержимое и ID пользователя (информация вводится с помощью input()), выполняет POST-запрос для создания нового поста и возвращает информацию о созданном посте в формате JSON


```
Заголовок: Новый пост
Содержимое поста: Тут должно находиться содержимое нового поста...
ID пользователя: 10
```



2. Реализуйте функцию, которая принимает ID поста, новый заголовок и новое содержимое, выполняет PUT-запрос и возвращает обновлённый пост в формате JSON

3. Реализуйте функцию, которая принимает ID поста, выполняет DELETE-запрос и возвращает статус-код ответа

````python
import requests


# 1) Функция для создания нового поста
def create_post(title, body, user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    payload = {
        "title": title,
        "body": body,
        "userId": user_id
    }
    response = requests.post(url, json=payload)
    post = response.json()
    return post

# 2) Функция для обновления поста
def update_post(post_id, new_title, new_body):
    url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
    payload = {
        "title": new_title,
        "body": new_body
    }
    response = requests.put(url, json=payload)
    updated_post = response.json()
    return updated_post

# 3) Функция для удаления поста
def delete_post(post_id):
    url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
    response = requests.delete(url)
    return response.status_code

# Главная функция
def main():
    functions = ['Создание нового поста', 'Обновление поста', 'Удаление поста', 'Выход из программы']

    while True:
        print('\nФункции программы: ')
        for count, function in enumerate(functions):
            print(f'{count + 1}) {function}')

        request = int(input('\nВведите номер функции: '))

        if request == 1:
            # Создание нового поста
            title = input("\nЗаголовок: ")
            body = input("Содержимое поста: ")
            user_id = int(input("ID пользователя: "))
            new_post = create_post(title, body, user_id)
            print("\nНовый пост:")
            print(new_post)

        if request == 2:
            # Обновление поста
            post_id = int(input("\nВведите ID поста для обновления: "))
            new_title = input("Новый заголовок: ")
            new_body = input("Новое содержимое: ")
            updated_post = update_post(post_id, new_title, new_body)
            print("\nОбновленный пост:")
            print(updated_post)

        if request == 3:
            # Удаление поста
            post_id = int(input("\nВведите ID поста для удаления: "))
            status_code = delete_post(post_id)
            print(f"Статус-код ответа: {status_code}")

        if request == 4:
          break

main()
````

## Задание 5. HTTP-запросы, ответы и пёсики

**Описание**

Создайте программу, которая будет взаимодействовать с Dog API, которая позволит получать список пород собак, вводить несколько пород и получать их фотогрфии.

Этапы:

1. Создайте функцию, которая использует метод GET и возвращает список всех пород собак в формате нумерованного списка

2. Реализуйте возможность ввода нескольких пород собак через запятую


```
african, chow, dingo
```



3. Создание функции, которая реализует запрос, возвращает и выводит изображениия собак, породы которых были введены до этого


Используйте Dog API (https://dog.ceo/dog-api/), который предоставляет информацию о породах собак и их изображения.


````python
import requests


# 1) Функция для получения списка всех пород собак
def get_breed_list():
    url = "https://dog.ceo/api/breeds/list/all"
    response = requests.get(url)
    data = response.json()
    breed_list = list(data['message'].keys())
    print("Список пород собак:")
    for count, breed in enumerate(breed_list, 1):
      print(f"{count}) {breed}")

# 2) Ввод пород собак через запятую
def get_breed_input():
    breed_input = input("\nВведите породы собак через запятую: ")
    breed_names = breed_input.split(", ")
    return breed_names

# 3) Функция для получения изображений собак по породам
def get_breed_images(breed_names):
    url = "https://dog.ceo/api/breed/{}/images"
    for breed in breed_names:
        breed_url = url.format(breed)
        response = requests.get(breed_url)
        data = response.json()
        if "message" in data:
            print(f"\nИзображения для породы '{breed}':")
            for image_url in data["message"]:
                print(image_url)
        else:
            print(f"Не удалось найти породу '{breed}'")

# 4) Главная функция
def main():
    get_breed_list()
    breed_names = get_breed_input()
    get_breed_images(breed_names)

main()
````
