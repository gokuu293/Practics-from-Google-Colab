# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

Дано:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
s = [int(i) for i in input().split()]

a = 0
for i in s:
    a  = a + i

mx = -10000
for  i in s:
    if i > mx:
        mx = i

k = 0
for i in s:
  k += 1

print(a)
print(a/k)
print(mx)
````

# **Задача 3**

Простой чат-бот
Создайте простого чат-бота, который отвечает на основные запросы пользователей. Программа должна использовать условные выражения для определения намерений пользователя и последующего соответствующего ответа.

Чат бот может:

- помогать складывать числа
- подсказывать фильм, музыку и тд (из заранее созданного списка)
- создавать список дел (а также выводить его, обновлять, очищать)

Интерфейс чат-бота на ваше усмотрение, это могут быть как заранее определенные команды, так и к примеру вызов команды по номеру

*Реализовать без использования функций*

Пример:

Ввод:

```
Олег, какую мызыку посоветуешь?
```

Вывод:

```
Я бы посоветовал вам: Led Zeppelin Stairway to Heaven
```

````python
commands = ['Какую музыку посоветуешь?', 'Какой фильм посоветуешь?', 'Помоги сложить числа', 'Покажи список моих дел',  \
            'Добавь новое дело в список дел', 'Очисти список моих дел', 'Выключить бота']

muz = ['Веселая музыка', 'Грустная музыка']
film = ['Веселый фильм', 'Грустный фильм']

works = []
k = 1

print('Список команд:')
for command in commands:
  print(f'{k}) {command}')
  k += 1

print('Введите команду из списка')

while True:
    a = input()

    if a == 'Какую музыку посоветуешь?':
      print('Какое у вас настроение: Веселое или Грустное?')
      a1 = str(input())
      if a1 == 'Веселое':
          print(muz[0])
      if a1 == 'Грустное':
          print(muz[1])

    if a == 'Какой фильм посоветуешь?':
        print('Какое у вас настроение: Веселое или Грустное?')
        a2 = str(input())
        if a2 == 'Веселое':
          print(film[0])
        if a2 == 'Грустное':
          print(film[1])

    if a == 'Помоги сложить числа':
      print('Введите количество чисел')
      c = 0
      k = 1
      count = int(input())
      for i in range(count):
        print('Введите' , k, 'число')
        i = int(input())
        c += i
        k += 1
      print('Сумма чисел равна', c)

    if a == 'Покажи список моих дел':
      if works != []:
        print(works)

      else:
        print('У вас нет дел')

    if a == 'Добавь новое дело в список дел':
      work = input('Введите дело')
      works.append(work)

    if a == 'Очисти список моих дел':
      works = []

    if a == 'Выключить бота':
      break

    else:
      print('Введите команду')




````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Вввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random


win_user = win_ai = 0

while True:
    if win_user == 3 or win_ai == 3:
      if win_user > win_ai:
        print(f'\nПобедил Игрок со счётом {win_user}:{win_ai} !')
      else:
        print(f'\nПобедил Компьютер со счётом {win_ai}:{win_user} !')
      break

    else:
      print(f'\nТекущий счёт: Игрок - {win_user}, Компьютер - {win_ai}')
      user_actions = input('Введите свой ход: ')

      actions = ['Ножницы', 'Камень' , 'Бумага']

      ai_actions = random.choice(actions)

      print(f'Выбор Компьютера: {ai_actions}')
      print(f'Выбор Игрока: {user_actions}')

      if user_actions == ai_actions:
        print('Исход: Ничья')

      elif user_actions == 'Бумага':
        if ai_actions == 'Камень':
          print('Победа: Игрок')
          win_user += 1
        else:
          print('Победа: Компьютер')
          win_ai += 1

      elif user_actions == 'Ножницы':
        if ai_actions == 'Камень':
          print('Победа: Компьютер')
          win_ai += 1
        else:
          print('Победа: Игрок')
          win_user += 1

      elif user_actions == 'Камень':
        if ai_actions == 'Ножницы':
          print('Победа: Игрок')
          win_user += 1
        else:
          print('Победа: Компьютер')
          win_ai += 1






````

Создать саму игру

````python
import random


guess_letters = []
points = 0

words = ['карандаш', 'ручка', 'линейка', 'пенал']
word = random.choice(words)
letter_word = [letter for letter in word]

print(f'Возможные слова: {words}')
int_try = int(input('Введите количество попыток: ')) + 1

while letter_word != [] and int_try != 1:
    int_try -= 1

    print(f'\nКоличество оставшихся попыток: {int_try},\
 Количество твоих очков: {points}' )

    print(f'Отгаданные буквы: {guess_letters}')

    gs = input('\n\nВведите предполагаемую букву в нижнем регистре на русском\
 языке: ')

    if gs in letter_word:
          print('Буква', gs, 'есть в слове')
          points += 1
          letter_word.remove(gs)
          guess_letters.append(gs)

    else:
        print('Буквы', gs, 'нет в слове или она уже была угадана,\
 либо введенный вами символ не является буквой')

if letter_word == []:
  print(f'\nВы угадали слово! Это {word}')

else:
  print(f'\nУ вас закончились попытки! Слово: {word}')

````
